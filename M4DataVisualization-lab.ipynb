# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_sql_query("SELECT ConvertedComp FROM master", conn)

plt.hist(df["ConvertedComp"], bins=50, alpha=0.7)
plt.title("Histogram of ConvertedComp")
plt.xlabel("ConvertedComp")
plt.ylabel("Frequency")
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_sql_query("SELECT Age FROM master", conn)

if not df.empty:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df["Age"])
    plt.title("Box Plot of Age")
    plt.xlabel("Age")
    plt.show()
else:
    print("The DataFrame is empty. Please check your database connection and query.")
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_sql_query("SELECT Age,WorkWeekHrs FROM master", conn)
plt.figure(figsize=(10,8))
plt.scatter(df["WorkWeekHrs"], df["Age"])
plt.title("Scatter Plot of Age vs WorkWeekHrs")
plt.xlabel("WorkWeekHrs")
plt.ylabel("Age")
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_sql_query("SELECT WorkWeekHrs,CodeRevHrs,Age FROM master", conn)
plt.figure(figsize=(10,8))
plt.scatter(df["WorkWeekHrs"], df["CodeRevHrs"], s=df["Age"]*5, alpha=0.7, color="skyblue", edgecolor="black")
plt.title("Bubble Plot of WorkWeekHrs, CodeRevHrs, and Age")
plt.xlabel("WorkWeekHrs")
plt.ylabel("CodeRevHrs")
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt 

QUERY = """
SELECT DatabaseDesireNextYear, Count(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""

df = pd.read_sql_query(QUERY, conn)

df.set_index('DatabaseDesireNextYear', inplace=True)


color_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
labels = df.index

plt.figure(figsize=(10, 6))
plt.pie(df['Count'], labels=labels, colors=color_list, autopct='%1.1f%%', startangle=90, shadow=True, pctdistance=0.85)
plt.title('Top 5 Database Desire Next Year') 
plt.axis('equal')  

plt.legend(labels, loc='upper left')

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_sql_query("SELECT Age, WorkWeekHrs, CodeRevHrs FROM master", conn)

age_group_df = df[(df["Age"] >= 30) & (df["Age"] <= 35)]

median_WorkWeekHrs = age_group_df["WorkWeekHrs"].median()
median_CodeRevHrs = age_group_df["CodeRevHrs"].median()

labels = ['WorkWeekHrs', 'CodeRevHrs']
sizes = [median_WorkWeekHrs, median_CodeRevHrs]

plt.bar(0, sizes[0], label=labels[0])
plt.bar(0, sizes[1], bottom=sizes[0], label=labels[1])
plt.xticks([0], ['Age Group 30-35'])
plt.title("Stacked Chart of Median WorkWeekHrs and CodeRevHrs")
plt.xlabel("Age Group")
plt.ylabel("Median WorkWeekHrs/CodeRevHrs")
plt.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_sql_query("SELECT Age, ConvertedComp FROM master", conn)


age_range_df = df[(df["Age"] >= 45) & (df["Age"] <= 60)]


median_ConvertedComp = age_range_df.groupby("Age")["ConvertedComp"].median()

plt.figure(figsize=(10,6))
median_ConvertedComp.plot(kind="bar")
plt.title("Median ConvertedComp by Age (45-60)")
plt.xlabel("Age")
plt.ylabel("Median ConvertedComp")
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_sql_query("SELECT MainBranch FROM master", conn)

branch_counts = df["MainBranch"].value_counts()

plt.figure(figsize=(10, 6))
plt.barh(branch_counts.index, branch_counts.values)
plt.title("Horizontal Bar Chart")
plt.xlabel("Count")
plt.ylabel("MainBranch")
plt.grid(True)
plt.show()
```
